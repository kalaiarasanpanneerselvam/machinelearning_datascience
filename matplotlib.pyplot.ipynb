## Different Option for Creating and Ploting the data

## Method1

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.set(title="My First Plot",
      xlabel="X-axis", ylabel="Y-axis",
      xlim=(0,10),ylim=(0,10))
x = [1,2,3,4,5]
y = [2,4,6,8,10]
ax.plot(x,y)
plt.show()
```

## Method2

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.set_title("My First Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("y-axis")
ax.set_xlim([0,10])
ax.set_ylim([0,10])
x = [1,2,3,4,5]
y = [2,4,6,8,10]
ax.plot(x,y)
plt.show()
```

## Method 3

```python
fig = plt.figure(figsize=(6,6))

plt.title("My First Plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.xlim([0,10])
plt.ylim([0,10])
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

# Types of Graphs 

### 1 Line Graph 

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.set(title="My First Plot",
      xlabel="X-axis", ylabel="Y-axis",
      xlim=(0,10),ylim=(0,10))
x = [1,2,3,4,5]
y = [2,4,6,8,10]
ax.plot(x,y, color='green', linestyle='--', linewidth=3, marker="o")
plt.show()
```

## 2 Bar Graph

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.set(title="My First Plot",
      xlabel="X-axis", ylabel="Y-axis",
      xlim=(0,10),ylim=(0,10))
x = [1,2,3,4,5]
y = [2,4,6,8,10]
ax.bar(x,y,color='red',width=.7,edgecolor='black')
plt.show()
```

## Horizontal Bar

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.set(title="My First Plot",
      xlabel="X-axis", ylabel="Y-axis",
      xlim=(0,10),ylim=(0,10))
x = [1,2,3,4,5]
y = [2,4,6,8,10]
ax.barh(x,y,color='green',edgecolor='black')
plt.show()
```

## Scatter Plot

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1,1,1)
ax.set(title="My First Plot",
      xlabel="X-axis", ylabel="Y-axis",
      xlim=(0,10),ylim=(0,10))
x = [1,2,3,4,5]
y = [2,4,6,8,10]
ax.scatter(x,y,color='green',marker='^')
plt.show()
```

## Pie Graph

```python

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode,labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
print(x)
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```

## Histogram

```python
import numpy as np 
np.random.seed(100)
x = 60 + 10 * np.random.randn(1000)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ax.set(title="Distribution of Student's Percentage",
      ylabel = "Count", xlabel = "Percentage" )
ax.hist(x)
plt.show()
print(x )
```

```python
N_points = 100000
n_bins = 20

# Generate a normal distribution, center at x=0 and y=5
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
```

## GridSpec.

```python
import matplotlib.gridspec as gridspec
fig = plt.figure()
gs = gridspec.GridSpec(3, 3)
ax1 = plt.subplot(gs[0, :])
ax2 = plt.subplot(gs[1, 0:1])
ax3 = plt.subplot(gs[1, 1:2])
ax4 = plt.subplot(gs[1, 2:3])
ax5 = plt.subplot(gs[2, :])

ax5.set(title="My First Plot",
      xlabel="X-axis", ylabel="Y-axis",
      xlim=(0,10),ylim=(0,10))
x = [1,2,3,4,5]
y = [2,4,6,8,10]
ax5.plot(x,y, color='green', linestyle='--', linewidth=3, marker="o")

plt.show()
```
